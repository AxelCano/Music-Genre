```python
import ast
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('tracks.csv')
df
```

```python
df.columns
```

```python
df[['track_genre_top','track_genres','track_genres_all']]
```

```python
def extract_numbers(x):
    return ast.literal_eval(x)

y = df['track_genres'].apply(extract_numbers)
```

```python
for i in range(0,len(y)):
    if len(y[i])>8:
        print(y[i])
```

```python
df['track_genre_top'].value_counts()
```

```python
plt.figure(figsize=(10,8))
plt.pie(df['track_genre_top'].value_counts(), labels = df['track_genre_top'].value_counts().index, autopct='%1.1f%%')
plt.title('Distribution of Genres')
plt.show()
```

```python
df['track_genre_list'] = df['track_genres'].apply(lambda x: ast.literal_eval(x))
all_tags = df['track_genre_list'].explode().value_counts()
all_tags.head(30)
```

```python


plt.figure(figsize=(10,6))

all_tags.head(30).plot(kind='bar', color='skyblue')

plt.title('Top 30', fontsize=16)
plt.xlabel('Genre', fontsize=12)
plt.ylabel('F', fontsize=12)

plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
